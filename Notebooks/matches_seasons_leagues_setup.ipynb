```python
for key, values in sets.items():
    for value in values:
        fp = 'statsbomb-matches/' + str(key) + '/' + str(value) + '.json'
        with open(fp, 'r') as f:
            test = json.load(f)
        df = pd.DataFrame(test)
        if "competition" in df.columns and "home_team" in df.columns and "away_team" in df.columns:
            match_ids = df.get("match_id")
            comp_names = df.get("competition").transform(lambda x: x['competition_name'])
            team_1_names = df.get("home_team").transform(lambda x: x['home_team_name'])
            team_2_names = df.get("away_team").transform(lambda x: x['away_team_name'])
            match_1_df = pd.DataFrame({"competition_names": comp_names, "team": team_1_names})
            match_2_df = pd.DataFrame({"competition_names": comp_names, "team": team_2_names})
            match_df = pd.concat([match_1_df, match_2_df])
            match_df["match_id"] = match_ids
            dfs.append(match_df)
        else:
            print(key, value)
```
