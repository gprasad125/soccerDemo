```python
import numpy as np
import pandas as pd
```

```python
nonNull = pd.read_csv('nonNull.csv')
nonNull.head(5)
```

```python
reduced = nonNull[['match_id', 'type_name']]
reduced
```

```python
coord_events = pd.read_csv('coord_events.csv')
coord_events
```

```python
combined = pd.concat([reduced, coord_events], axis = 1)
combined
```

```python
locs = combined.get("match_id") == 3749552
```

```python
combined.loc[locs, :] = combined.loc[locs].replace({'Manchester United': 'Manchester United Men'})



# 4 matches need to change - (3749552, 3749246, 18236, 3750201)
```

```python
# cleaning up women's soccer teams - two have changed names in the last 3/4 years
```
