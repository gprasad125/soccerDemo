```python
passes_players = pd.read_csv('passes_players.csv')
passes_players.head(5)
```

```python
sheridan = passes_players[passes_players.get('player_name').transform(lambda x: 'Sheridan' in x)]
sheridan
```

```python
sheridan['player_name'] = sheridan['player_name'].replace({'Kailen Mary Iacovoni Sheridan': 'Kailen Sheridan'})
sheridan
```

```python
sher_reduced = sheridan.groupby('player_name').mean()[['location_x', 'location_y', 'pass_length', 'pass_angle', 'Success']]
```

```python
sher_reduced = sher_reduced.rename(columns = {'Success': 'simple_pass_accuracy'})
sher_reduced['team_name'] = 'NJ/NY Gotham FC'
sher_reduced['position_name'] = 'Goalkeeper'
sher_reduced = sher_reduced.reset_index()
sher_reduced
```

```python
locs = passes_players.get("match_id") == 3749133
passes_players.loc[locs, :] = passes_players.loc[locs].replace({'Aston Villa': 'Aston Villa Men'})

locs = passes_players.get("match_id") == 3749552
passes_players.loc[locs, :] = passes_players.loc[locs].replace({'Manchester United': 'Manchester United Men'})

locs = passes_players.get("match_id") == 3749246
passes_players.loc[locs, :] = passes_players.loc[locs].replace({'Manchester United': 'Manchester United Men'})

locs = passes_players.get("match_id") == 18236
passes_players.loc[locs, :] = passes_players.loc[locs].replace({'Manchester United': 'Manchester United Men'})

locs = passes_players.get("match_id") == 3750201
passes_players.loc[locs, :] = passes_players.loc[locs].replace({'Manchester United': 'Manchester United Men'})

passes_players['team_name'] = passes_players['team_name'].replace({'Seattle Reign': 'OL Reign', 'Sky Blue FC': 'NJ/NY Gotham FC'})
```

```python
total = merged_reset.merge(passes_players, left_index = True, right_index = True)
total.head(5)
```

```python
total.columns
```

```python
total = total.drop(['type_name_y', 'team_name_x', 'player_name_y'], axis = 1).rename(columns = {'type_name_x': 'type_name', 'team_name_y': 'team_name', 'player_name_x': 'player_name'})
total.head(5)
```

```python
reduced = total[['team_name', 'player_name', 'position_name', 'location_x', 'location_y', 'Success', 'pass_length', 'pass_angle']]
reduced.head(5)
```

```python
grouped = reduced.groupby(['team_name', 'position_name', 'player_name']).mean()
grouped
```

```python
grouped = grouped.rename(columns = {'Success': 'simple_pass_accuracy'})
grouped.head(5)
```

```python
grouped.to_csv('player_pos_averages.csv', index = True)
```
